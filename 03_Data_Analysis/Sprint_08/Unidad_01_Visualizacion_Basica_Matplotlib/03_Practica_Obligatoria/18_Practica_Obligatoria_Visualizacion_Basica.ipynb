#### #1.2  

Recrea la siguiente figura, para ello construye en apartados separados cada grfico y luego haz un apartado en el que se junten los cuatro.  

Las figuras son:
1. Las ciudades ms pobladas, con los numeros de las poblaciones (tendrs que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrn, con trasparencia)  

2. Histograma del rea total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand ITC", tamao 18, color azul acero o "steelblue")
3. El scatter de rea de agua de un ciudad con su poblacion.
4. El boxplot de la altura de las ciudades.

```python
# ciudades ms pobladas
df.sort_values(by = "population_total", ascending = False)[0:5]
```

```python
ciudades = ["Los Angeles", "San Diego", "Sacramento", "Anaheim", "Santa Ana"]
frecuencias = (df["population_total"]/1000).sort_values(ascending = False)[0:5].value_counts().index.to_list()

plt.bar(ciudades, frecuencias, color = ["blue", "grey", "red", "olive", "brown"], alpha = 0.5)
plt.title("Ciudades ms pobladas de California")
plt.xlabel("Ciudades")
plt.ylabel("Habitantes (miles)")
plt.yticks(range(0, 5001, 1000))

for ciudad, frec in zip(ciudades, frecuencias):
    plt.text(ciudad, frec + 100, str(frec), ha = "center")
```

```python
# Histograma del rea total
font_dict = {
    "size": 18,
    "family": "Bradley Hand ITC"
}
plt.hist(df["area_total_km2"], bins = 50, color = "steelblue", label = "Hist Area")
plt.xlabel("Superficie ($km^2$)", fontdict = font_dict)
plt.ylabel("Nmero de ciudades", fontdict = font_dict)
plt.legend();
```

```python
# scatter de rea de agua
outlier_1_x = df["area_water_km2"].max()
outlier_1_y = df.loc[df["area_water_km2"] == outlier_1_x, "population_total"].values[0]
outlier_2_y = df["population_total"].max()
outlier_2_x = df.loc[df["population_total"] == outlier_2_y, "area_water_km2"].values[0]

plt.scatter(df["area_water_km2"], df["population_total"]/1000, c = "green", alpha = 0.5)
plt.title("Poblacin por kilmetro cuadrado de agua")
plt.xlabel("rea de agua en la ciudad ($km^2$)")
plt.ylabel("Poblacin (miles habitantes)")
plt.annotate("Outliers", xytext = (60, 3500), xy = (outlier_1_x, outlier_1_y/1000), arrowprops = {"arrowstyle": "->"})
plt.annotate("Outliers", xytext = (60, 3500), xy = (outlier_2_x, outlier_2_y/1000), arrowprops = {"arrowstyle": "->"});
```

```python
# boxplot de altura de las ciudades
plt.boxplot(df["elevation_m"], vert = False, patch_artist = True, boxprops = {"facecolor": "red"}, medianprops = {"color": "k"})
plt.title("Diagrama de caja de la Altitud de las ciudades de California")
plt.xlabel("Altitud (m)");
```

```python
fig = plt.figure(figsize = (15, 15))

# ciudades ms pobladas
plt.subplot(2, 2, 1)
ciudades = ["Los Angeles", "San Diego", "Sacramento", "Anaheim", "Santa Ana"]
frecuencias = (df["population_total"]/1000).sort_values(ascending = False)[0:5].value_counts().index.to_list()

plt.bar(ciudades, frecuencias, color = ["blue", "grey", "red", "olive", "brown"], alpha = 0.5)
plt.title("Ciudades ms pobladas de California")
plt.xlabel("Ciudades")
plt.ylabel("Habitantes (miles)")
plt.yticks(range(0, 5001, 1000))

for ciudad, frec in zip(ciudades, frecuencias):
    plt.text(ciudad, frec + 100, str(frec), ha = "center")

# Histograma del rea total
plt.subplot(2, 2, 2)
font_dict = {
    "size": 18,
    "family": "Bradley Hand ITC"
}
plt.hist(df["area_total_km2"], bins = 50, color = "steelblue", label = "Hist Area")
plt.xlabel("Superficie ($km^2$)", fontdict = font_dict)
plt.ylabel("Nmero de ciudades", fontdict = font_dict)
plt.legend();

# scatter de rea de agua
plt.subplot(2, 2, 3)
outlier_1_x = df["area_water_km2"].max()
outlier_1_y = df.loc[df["area_water_km2"] == outlier_1_x, "population_total"].values[0]
outlier_2_y = df["population_total"].max()
outlier_2_x = df.loc[df["population_total"] == outlier_2_y, "area_water_km2"].values[0]

plt.scatter(df["area_water_km2"], df["population_total"]/1000, c = "green", alpha = 0.5)
plt.title("Poblacin por kilmetro cuadrado de agua")
plt.xlabel("rea de agua en la ciudad ($km^2$)")
plt.ylabel("Poblacin (miles habitantes)")
plt.annotate("Outliers", xytext = (60, 3500), xy = (outlier_1_x, outlier_1_y/1000), arrowprops = {"arrowstyle": "->"})
plt.annotate("Outliers", xytext = (60, 3500), xy = (outlier_2_x, outlier_2_y/1000), arrowprops = {"arrowstyle": "->"})

# boxplot de altura de las ciudades
plt.subplot(2, 2, 4)
plt.boxplot(df["elevation_m"], vert = False, patch_artist = True, boxprops = {"facecolor": "red"}, medianprops = {"color": "k"})
plt.title("Diagrama de caja de la Altitud de las ciudades de California")
plt.xlabel("Altitud (m)");
```
